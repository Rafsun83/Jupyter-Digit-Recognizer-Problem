```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
```

```python
model=MLPClassifier(hidden_layer_sizes=(500,500,500 ),beta_1=0.2, batch_size=250, shuffle=True, learning_rate_init=0.0005, tol=0.00001, max_iter=50, early_stopping=True,verbose=1, random_state=100)
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score=cross_val_score(model,X,y,cv=5)
```

```python
cv_score
```

```python
cv_score.mean()
```
